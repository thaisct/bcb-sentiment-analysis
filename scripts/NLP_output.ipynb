```python
fig = go.Figure(layout = go.Layout(width=1400, height=700))
fig.add_trace(go.Scatter(x = list(df_finbert_s.index), y= flatten(np.ndarray.tolist(df_finbert_s.values)),
                         name="MÃ©dia MÃ³vel da SÃ©rie Original (FinBERT)", line = dict(color = 'blue')))
fig.add_trace(go.Scatter(x = list(df_f2_s.index), y= flatten(np.ndarray.tolist(df_f2_s.values)),
                         name="MÃ©dia MÃ³vel (FinBERT refinado)", line = dict(color = 'blue', dash = 'dash')))
fig.add_trace(go.Scatter(x = list(df_uncased_s.index), y= flatten(np.ndarray.tolist(df_uncased_s.values)),
                         name="MÃ©dia MÃ³vel da SÃ©rie Original (BERT uncased)", line = dict(color = 'red')))
fig.add_trace(go.Scatter(x = list(df_u1_s.index), y= flatten(np.ndarray.tolist(df_u1_s.values)),
                         name="MÃ©dia MÃ³vel (BERT uncased refinado)", line = dict(color = 'red', dash = 'dash')))

fig.update_layout(
    title=dict(text = "Sentimento das Atas", font=dict(size=40)),
    xaxis_title="Datas de PublicaÃ§Ã£o",
    yaxis_title="Sentimento",
    legend_title="",
        legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.13,
        bgcolor = "rgba(0,0,0,0)"
        ),
    font = dict(size = 15)
    )

periods = [
    dict(start='2007-07-01', end='2009-06-30', mid_x = '2008-07-01', mid_y = 18, name = 'GFC' ,color='orange'),  
    dict(start='2015-11-01', end='2016-11-30', mid_x = '2016-05-30', mid_y = 18,name = 'Crise PolÃ­tica', color='orange'),
    dict(start='2018-05-01', end='2018-06-30', mid_x = '2018-05-31', mid_y = 25,name = 'Greve dos Caminhoneiros', color='orange'),
    dict(start='2020-04-01', end='2021-10-31', mid_x = '2021-01-15', mid_y = 18,name = 'Pandemia', color='orange'),  
]

# Add the rectangles to the layout
for period in periods:
    fig.add_shape(
        type="rect",
        xref="x",
        yref="paper",
        x0=period['start'],
        y0=0,
        x1=period['end'],
        y1=1,
        fillcolor=period['color'],
        opacity=0.5,
        layer="below",
        line_width=0,
    )
        # Add the annotation for the period name
    fig.add_annotation(
        x=period['mid_x'],
        y=period['mid_y'],
        text=period['name'],
        showarrow=False,
        font=dict(size=12, color='black'),
    )
fig.show()
```
