```python
import pandas as pd
import numpy as np
import matplotlib as plt
import plotly.express as px 
import plotly.graph_objects as go
import os
os.chdir("C:\\Users\\Thais\\Documents\\Python\\bcb-sentiment-analysis")
```

```python
def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
df_original = pd.DataFrame(pd.read_pickle(".\\data\\minutes\\minutes_scores_original.pkl", compression='infer', storage_options=None).set_index('date')['score'][8:])
df_original_sm = pd.DataFrame(pd.read_pickle(".\\data\\minutes\\minutes_scores_original.pkl", compression='infer', storage_options=None).set_index('date')['score'][8:].rolling(window=4).mean())
df_s1 = pd.DataFrame(pd.read_pickle(".\\data\\minutes\\minutes_scores_finbert_20230423.pkl", compression='infer', storage_options=None).set_index('date')['score'][8:].rolling(window=4).mean())
df_s2 = pd.DataFrame(pd.read_pickle(".\\data\\minutes\\minutes_scores_finbert_20230430.pkl", compression='infer', storage_options=None).set_index('date')['score'][8:].rolling(window=4).mean())
df_s3 = pd.DataFrame(pd.read_pickle(".\\data\\minutes\\minutes_scores_uncased_20230423.pkl", compression='infer', storage_options=None).set_index('date')['score'][8:].rolling(window=4).mean())
df_s4 = pd.DataFrame(pd.read_pickle(".\\data\\minutes\\minutes_scores_uncased_20230430.pkl", compression='infer', storage_options=None).set_index('date')['score'][8:].rolling(window=4).mean())

df_s1.columns = ['score_f1']
df_s2.columns = ['score_f2']
df_s3.columns = ['score_u1']
df_s4.columns = ['score_u2']
```

# Fig 1

```python
fig = go.Figure(layout = go.Layout(width=1400, height=700))
fig.add_trace(go.Scatter(x = list(df_original.index), y= flatten(np.ndarray.tolist(df_original.values)),
                         name="Série Original"))
fig.add_trace(go.Scatter(x = list(df_original_sm.index), y= flatten(np.ndarray.tolist(df_original_sm.values)),
                         name="Média Móvel de 3 reuniões"))
fig.update_layout(
    title=dict(text = "Sentimento das Atas (FinBERT)", font=dict(size=40)),
    xaxis_title="Datas de Publicação",
    yaxis_title="Sentimento",
    legend_title="",
        legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.8,
        bgcolor = "rgba(0,0,0,0)"
        ),
    font = dict(size = 15)
    )
fig.show()
```
